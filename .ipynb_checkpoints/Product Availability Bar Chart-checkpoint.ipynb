```python
import pandas as pd
```

```python
#read data
chewy_df=pd.read_csv('chewy_review.csv')
```

```python
#Prepare data for visualization of the top 10 brands with most product variety on chewy website
brand=chewy_df[['brandName', 'productName']].copy()
brand.drop_duplicates(keep='first',inplace=True) 
brand['Count_brandName']=1
brandGroupdf=brand.groupby('brandName').sum()

Top10=brandGroupdf.sort_values(by='Count_brandName',ascending=False).head(10)
```

```python
#Use matplotly to make the graphs
Top10.plot.bar()

plt.title('Dog Food: Product Counts for each Brand')
plt.xlabel('Brand')
plt.ylabel('Count of Products per Brand')
```

```python
#use seaborn to make graph
import seaborn as sns
%matplotlib inline

sns.set(style="whitegrid")
chart=sns.barplot(x=Top10.index, y="Count_brandName", data=Top10, )

plt.xlabel('Brand')
plt.ylabel('Number of Products')
plt.title('Product Availability on Chewy.com (Top 10 Brands)')
chart.set_xticklabels(chart.get_xticklabels(), rotation=90)
```
